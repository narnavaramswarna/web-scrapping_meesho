taking list and appending all srcs to this list (filtering imgs srcs)

```python
srcs=[]
for i in soup.find_all('img'):
    k= i.get('src')
    if k.startswith('http'):
        srcs.append(k)
         
srcs    
```

importing urlparse for detecting type of src file means(.jpg,.svg,.png) etc

```python
from urllib.parse import urlparse
```

Main code:
i am iterating all links in srcs with i,j -->i=index of image,j-->link of img
then using requests taking the url in j and stored in res
checking status of res object is it okay successfully fetched link
now checking extenstion of link using urlparse-->j
os.path.splitext()=splits the file 2 paths
ex:(root,extension)
root:takes everything except extension
extension=.
[1]=>means take only last extension 
ex:"images_watch/photo_archive.tar.gz"-->takes only ".gz"
if this fetching is empty by default i am storing it as ".jpg
now assigning path of  downloded images to dis

opening this dis file in wb mode(images->binary images) as file
write on file res.content
after print ("sucess)
else:
display("unscuess")

```python
for i,j in enumerate(srcs):
    res=requests.get(j,stream=True)
    if res.status_code==200:
        url_type=urlparse(j)
        fetching=os.path.splitext(url_type.path)[1]
        if not fetching:
            fetching=".jpg"
        dis=os.path.join(image_folder,f'image{i}.{fetching}')    
        with open(dis,'wb') as file:
            file.write(res.content)
        print(f"Downloaded:{dis}")
    else:
        print("unsceesfull {j} url")        
```
